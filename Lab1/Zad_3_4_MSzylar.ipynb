Opis ćwiczenia - zadanie 3:
W zadaniu należy zdefiniować funkcję f(x)=x^2+5 na poniższych trzech przedziałach:

1.x>-1 oraz x<1

2.x>-6 oraz x<6

3.x>0 oraz x<5

Wspierając się dokumentacją Matplotlib. Dodać do wykresu etykiety osi, tytuły wykresów i legendy.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def fun(x):
    return x*x+5
```

```python
x1 = np.linspace(-1,1,100)
x2 = np.linspace(-6,6,100)
x3 = np.linspace(-0,5,100)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
fig.suptitle('Wykresy funkcji w zależności od przedziału argumentów')
ax1.plot(x1, fun(x1), label="od -1 do 1")
ax1.set_xlabel('Argument')
ax1.set_ylabel('Wartość')
ax1.legend(loc='upper left')
ax2.set_xlabel('Argument')
ax2.set_ylabel('Wartość')
ax2.plot(x2, fun(x2), label="od -6 do 6")
ax2.legend(loc='upper left')
ax3.plot(x3, fun(x3), label="od 0 do 5")
ax3.set_xlabel('Argument')
ax3.set_ylabel('Wartość')
ax3.legend(loc='upper left')
```

Wszystkie polecenia zawarte w opisie ćwiczenia zostały zrealizowane.
